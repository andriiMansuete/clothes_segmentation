```python
BACKBONE = 'efficientnetb3'
# BATCH_SIZE = 8
# CLASSES = all_CLASSES
# LR = 0.0001
# EPOCHS = 50

preprocess_input = sm.get_preprocessing(BACKBONE)
# define network parameters
n_classes = len(config['all_classes'])
activation = 'sigmoid' if n_classes == 1 else 'softmax'

#create model
model = sm.Unet(BACKBONE, classes=n_classes, activation=activation)
model.load_weights(config['model'])
```

```python
img = cv2.imread('./preprocess_data/data/custom_test/26-fendi-women-fw-21-22-d96f51.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def denormalize(x):
    """Scale image to range 0..1 for correct plot"""
    x_max = np.percentile(x, 98)
    x_min = np.percentile(x, 2)    
    x = (x - x_min) / (x_max - x_min)
    x = x.clip(0, 1)
    return x


def get_custom_test_augmentation():
    """Add paddings to make image shape divisible by 32"""
    test_transform = [
        A.Resize(384, 480),
        A.PadIfNeeded(384, 480)
    ]
    return A.Compose(test_transform)


def get_preprocessing(preprocessing_fn):
    """Construct preprocessing transform

    Args:
        preprocessing_fn (callbale): data normalization function 
            (can be specific for each pretrained neural network)
    Return:
        transform: albumentations.Compose

    """

    _transform = [
        A.Lambda(image=preprocessing_fn),
    ]
    return A.Compose(_transform)
```

```python
img.shape
```

```python
agumentation = get_custom_test_augmentation()
sample_augmented = agumentation(image=img)
augmented_image = sample_augmented['image']
```

```python
augmented_image.shape
```

```python
preprocessing = get_preprocessing(preprocess_input)
sample_processed = preprocessing(image=augmented_image)
processed_image = sample_processed['image']
```

```python
processed_image.shape
```

```python
processed_image = np.expand_dims(processed_image, axis=0)
pr_mask = model.predict(processed_image)
```

```python
label_image = pr_mask.squeeze().argmax(axis=2)
```

```python
image_label_overlay = label2rgb(label_image, image=denormalize(processed_image.squeeze()), bg_label=0)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.imshow(image_label_overlay)

for region in regionprops(label_image):
    # take regions with large enough areas
    if region.area >= 500:
        # draw rectangle around segmented coins
        minr, minc, maxr, maxc = region.bbox
        rect = mpatches.Rectangle((minc, minr), maxc - minc, maxr - minr,
                                  fill=False, edgecolor='red', linewidth=2, label='Label')
        ax.add_patch(rect)
        rx, ry = rect.get_xy()
        cx = rx + rect.get_width()/2.0
        cy = ry + rect.get_height()/2.0
        ax.annotate(['bg', 'shirt, blouse', 'top, t-shirt, sweatshirt', 'sweater', 'cardigan',
       'jacket', 'pants', 'shorts', 'skirt', 'coat', 'dress', 'hat',
       'tie', 'glove', 'shoe'][region.label], (cx, cy), color='black', weight='bold', fontsize=10, ha='right', va='center')

ax.set_axis_off()
plt.tight_layout()
# plt.savefig("test.png")
plt.show()
```
